```python
import matplotlib.pyplot as plt # toliq import qilish va bu 'pyplot' graphic larni oson chizishga kere
import numpy as np # chart quramiz misol un
```

# Basic Line Plot

```python
x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 25, 30])

plt.plot(x, y)
plt.title('Basic Line Plot')
plt.xlabel('X Values')
plt.ylabel('Y value')
plt.show() # bu muhim method ! bu qoymasak "Text(0, 0.5, 'Y value')" ham korinyapti | clean visual un kerak
```

# Line Styles and Colors

```python
# yana bir misol
x = np.array([1, 2, 3, 4])
y = np.array([3, 8, 1, 10])

# x ni yozmasak VALUEni "0"dan boshlab yozib olyapti !!
plt.plot(y, linestyle='dotted', color='green', marker='o')
plt.title('Styled Line Example')
plt.show()
```

# Multiple Lines in One Graph with Numpy and Matplotlib

```python
# bir dona line emas, bir nechta LINES bolishi mumkin
x = np.array([1, 2, 3, 4])
y1 = np.array([2, 4, 6, 8])
y2 = np.array([1, 3, 5, 7])

plt.plot(x, y1, label='Line 1', color='red', marker='s')
plt.plot(x, y2, label='Line 2', color='blue', linestyle='--', marker='o')
plt.title('Multiple Lines Example')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.legend()
plt.show()
```

## **BAR CHART with NUMPY and MATPLOTLIB**

```python
# BAR CHART EXAMPLE

# x = np.array(['A', 'B', 'C', 'D'])
x = np.array(['Salary', 'Expense', 'Charity', 'Food'])
y = np.array([5, 9, 1, 7])

plt.bar(x, y, color='orange')
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

# Scatter Plot with Numpy and Matplotlib

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 4, 6, 3, 7])

plt.scatter(x, y, color='purple', s=100) # "s" used for SIZE !!!
plt.title('Scatter Plot Example')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

# Histogram with Numpy and Matplotlib

```python
data = np.array([3, 4, 5, 6, 8, 8, 8, 9, 10])

# "bins" ni qoymasak raqamlar yoq joylarni ochiq qoldiradi
# plt.hist(data, color='skyblue', edgecolor='black')
plt.hist(data, bins=5, color='skyblue', edgecolor='black') # bins bu bar larni miqdori
plt.title('Histogram Example')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.show()
```

# Subplot (Multiple Graphs Together)

```python
x = np.array([1, 2, 3, 4])
y1 = np.array([2, 4, 6, 7])
y2 = np.array([3, 6, 9, 12])

plt.figure(figsize=(8, 4))   # (width, height)

plt.subplot(1, 2, 1)  # bu raqamlarda "1-QATOR", "2-USTUN", "3-INDEX" ni anglatadi
plt.plot(x, y1, color='red')
plt.title('Plot 1')

plt.subplot(1, 2, 2)
plt.bar(x, y2, color='blue')
plt.title('Plot 2')

plt.tight_layout()
plt.show()
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([4, 8, 6, 10])

plt.plot(x, y, color='green')
plt.title('Save Figure Example')
plt.savefig('my_plot.png') # Saves in current folder when working locally !!!
plt.show()
```

```python
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import axes3d

plt.style.use('_mpl-gallery')

# Make data
X, Y, Z = axes3d.get_test_data(0.05)

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# Make data
np.random.seed(19680801)
n = 100
rng = np.random.default_rng()
xs = rng.uniform(23, 32, n)
ys = rng.uniform(0, 100, n)
zs = rng.uniform(-50, -25, n)

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(xs, ys, zs)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```
